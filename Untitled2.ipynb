```python
# Import library yang diperlukan
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
# Pastikan file `Car_sales.csv` ada di direktori kerja Anda
df = pd.read_csv('Car_sales.csv')

# Tampilkan info dataset
print("Informasi Dataset:")
print(df.info())

# Tampilkan statistik deskriptif
print("\nStatistik Deskriptif Dataset:")
print(df.describe())

# 1. Pembersihan Data
# Isi nilai kosong dengan median untuk kolom numerik
df['__year_resale_value'] = df['__year_resale_value'].fillna(df['__year_resale_value'].median())

# Hapus duplikasi
df = df.drop_duplicates()

# Tangani outlier pada 'Price_in_thousands' (opsional, jika diperlukan)
q1 = df['Price_in_thousands'].quantile(0.25)
q3 = df['Price_in_thousands'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
df = df[(df['Price_in_thousands'] >= lower_bound) & (df['Price_in_thousands'] <= upper_bound)]

# 2. Visualisasi Distribusi Data
# a. Distribusi penjualan mobil
plt.figure(figsize=(8, 5))
sns.histplot(df['Sales_in_thousands'], bins=20, kde=True)
plt.title('Distribusi Penjualan Mobil')
plt.xlabel('Penjualan (ribuan)')
plt.ylabel('Frekuensi')
plt.show()

# b. Harga mobil vs Penjualan
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Price_in_thousands', y='Sales_in_thousands', data=df, hue='Fuel_efficiency', palette='coolwarm')
plt.title('Harga Mobil vs Penjualan')
plt.xlabel('Harga (ribuan USD)')
plt.ylabel('Penjualan (ribuan unit)')
plt.show()

# c. Korelasi antar variabel
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriks Korelasi')
plt.show()

# 3. Analisis Lanjutan
# a. Penjualan berdasarkan kategori harga
bins = [0, 20, 40, 100]
labels = ['Low', 'Medium', 'High']
df['Price_Category'] = pd.cut(df['Price_in_thousands'], bins=bins, labels=labels)

sales_by_price = df.groupby('Price_Category')['Sales_in_thousands'].mean()
plt.figure(figsize=(8, 5))
sales_by_price.plot(kind='bar', color=['skyblue', 'orange', 'red'])
plt.title('Rata-rata Penjualan Berdasarkan Kategori Harga')
plt.xlabel('Kategori Harga')
plt.ylabel('Rata-rata Penjualan (ribuan unit)')
plt.show()

# b. Efisiensi bahan bakar vs Harga
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Fuel_efficiency', y='Price_in_thousands', data=df, hue='Price_Category', palette='viridis')
plt.title('Efisiensi Bahan Bakar vs Harga Mobil')
plt.xlabel('Efisiensi Bahan Bakar (MPG)')
plt.ylabel('Harga (ribuan USD)')
plt.show()

# c. Faktor performa vs Penjualan
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Power_perf_factor', y='Sales_in_thousands', data=df, hue='Price_Category', palette='coolwarm')
plt.title('Faktor Performa vs Penjualan')
plt.xlabel('Faktor Performa')
plt.ylabel('Penjualan (ribuan unit)')
plt.show()

# 4. Kesimpulan
print("\nKesimpulan:")
print("- Mobil dengan kategori harga rendah ('Low') memiliki rata-rata penjualan lebih tinggi dibanding kategori lainnya.")
print("- Efisiensi bahan bakar yang lebih tinggi cenderung ditemukan pada mobil dengan harga menengah.")
print("- Faktor performa menunjukkan mobil dengan performa tinggi lebih cocok untuk pasar niche dengan penjualan lebih kecil.")
```
